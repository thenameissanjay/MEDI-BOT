```python
# borrow from makeittalk
import ipywidgets as widgets
import glob
import matplotlib.pyplot as plt
print("Choose the image name to animate: (saved in folder 'examples/')")
img_list = glob.glob1('examples/source_image', 'abu.png')
img_list.sort()
img_list = [item.split('.')[0] for item in img_list]
default_head_name = widgets.Dropdown(options=img_list, value='abu')
def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        plt.imshow(plt.imread('examples/source_image/{}.png'.format(default_head_name.value)))
        plt.axis('off')
        plt.show()
default_head_name.observe(on_change)
display(default_head_name)
plt.imshow(plt.imread('examples/source_image/{}.png'.format(default_head_name.value)))
plt.axis('off')
plt.show()
```

```python
import gradio as gr
import os
import glob

# Define a function to load and display the video
def play_video():
    # Get the path to the directory containing the video
    video_directory = './results/2023_10_09_05.34.03/'

    # Check if the directory exists
    if os.path.exists(video_directory):
        # Search for the video file in the directory
        video_files = glob.glob(os.path.join(video_directory, 'abu##RD_Radio31_000_full.mp4'))

        if video_files:
            mp4_name = video_files[0]
            return gr.Video(mp4_name)
        else:
            return "No matching video file found in the directory."
    else:
        return "The specified directory does not exist."

# Create a Gradio interface
iface = gr.Interface(
    fn=play_video,
    inputs=None,  # No input component required
    outputs=gr.Video()  # Output is a Gradio Video component
)

# Launch the Gradio interface
iface.launch(share=True)
```
